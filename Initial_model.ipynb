```python
import pandas as pd
eth = pd.read_csv('Tweets-Dogecoin.csv')
eth
```

```python
import re
for b in range(len(eth['tweet'])):
  try:
    text = eth['tweet'][b]

    #filtering out links
    text = re.sub(r'https?://\S+|www\.\S+', '.', text)
    text = re.sub(r'\\n','',text)
    text = re.sub(r'\b[^\s]*\.com[^\s]*\b','',text)
    text = re.sub(r'\b[^\s]*\.net[^\s]*\b','',text)
    text = re.sub(r'\b[^\s]*\.org[^\s]*\b','',text)
    text = re.sub(r'\b[^\s]*\.gov[^\s]*\b','',text)


    text= re.sub(r'\.','. ',text) #converts ... or .., etc into .
    text = re.sub("[^A-Za-z0-9$,#%!\)\(‚Äî.;:'\"\&/ =\+-]","",text) #removes anything that is not within the specified characters
    text = re.sub(r' +',' ', text)
    text = text.strip()
    text = text.lower()
    eth['tweet'][b] = text
  except:
    print(b, eth['tweet'][b])
    eth = eth.drop(b)
eth.reset_index(inplace=True,drop = True)
```

```python
eth.to_csv('cleaned_tweets.csv')
```
